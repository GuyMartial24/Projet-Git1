```python
def energie(masse,hauteur, energie_limit, g=9.81):
    e= masse*hauteur*g
    if e<energie_limit:
        print(f"l'eneregie {e} est inférieure à l'eneregie limite {energie_limit}/n")

    else:
        print(f"l'eneregie {e} est supérieure à l'energie limite {energie_limit}")

    return(f"{e} Joules")

```

```python
energie(80,20,100,10)
```

```python
classeur={
    "positif":[],
    "negatif":[]

}
```

```python
def trier(classeur, nbr):
    if nbr<0:
        classeur["negatif"].append(nbr)
    else:
        classeur["positif"].append(nbr)
        return classeur
```

```python
trier(classeur,4)

print(classeur["positif"][0])

```
